```python
import glob
import pandas as pd
import warnings
warnings.filterwarnings("ignore") 
import datetime
```

```python
today=str(datetime.datetime.today().strftime('%d_%m_%Y'))
```

```python
INC = pd.read_excel(r'C:\NotBackedUp\SOE\source_files\IncidentCausedbyCHG\u_inc_change.xlsx')
```

```python
model1 = pd.read_csv(r'C:\NotBackedUp\SOE\source_files\service_mapping\model1_mapping.csv')
```

```python
model1 = model1.loc[model1.notnull().sum(1).groupby(model1.Name).idxmax()]
```

```python
SIAMMap = pd.read_csv(r'C:\NotBackedUp\SOE\source_files\service_mapping\SIAM_ServiceMapping.csv')
```

```python
SIAMMap = SIAMMap.loc[SIAMMap.notnull().sum(1).groupby(SIAMMap.Name).idxmax()]
```

```python
INC = INC.rename(columns={'Caused by Change':'CausedbyCHG'})
```

```python
INC['CausedbyCHG'] = INC['CausedbyCHG'].fillna(0)
```

```python
I = INC[INC['CausedbyCHG'] != 0]
```

```python
I
```

```python
I['Name'] = I['Name'].str.strip('[PROD]')
I['Name'] = I['Name'].str.strip()
```

```python
I = I.rename(columns={'Name':'business_service_name'})
```

```python
model1_dat = I.merge(model1,how='left',left_on=['business_service_name'],right_on=['Name'])
SIAM_dat = I.merge(SIAMMap,how='left',left_on=['business_service_name'],right_on=['Name'])
```

```python
model1_dat = model1_dat[['business_service_name','Number','CausedbyCHG','Domain_y','Tribe_TechArea','Name','Incident state']]
SIAM_dat = SIAM_dat[['business_service_name','Number','CausedbyCHG','Domain_y','Tribe_TechArea','Name','Incident state']]
```

```python
SIAM_dat['Name'] = SIAM_dat['Name'].fillna(0)
model1_dat['Name'] = model1_dat['Name'].fillna(0)
```

```python
SIAM_dat = SIAM_dat.rename(columns={'Domain_y':'Domain'})
model1_dat = model1_dat.rename(columns={'Domain_y':'Domain'})
```

```python
model1_dat = model1_dat[model1_dat['Name'] != 0]
SIAM_dat = SIAM_dat[SIAM_dat['Name'] != 0]
```
